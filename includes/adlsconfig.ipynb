```python
#creating the variables for secret scopes and incorporating keys with them to be used in Oauth authentication
clientid = dbutils.secrets.get(scope = 'f1scope', key= 'clientid')
tenantid = dbutils.secrets.get(scope ='f1scope', key = 'tenant')
secretid = dbutils.secrets.get(scope = 'f1scope', key = 'secretid')
# connecting databricks to azure using the above secrets
spark.conf.set(f"fs.azure.account.auth.type.formula1projectdb.dfs.core.windows.net", "OAuth")
spark.conf.set(f"fs.azure.account.oauth.provider.type.formula1projectdb.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set(f"fs.azure.account.oauth2.client.id.formula1projectdb.dfs.core.windows.net", clientid)
spark.conf.set(f"fs.azure.account.oauth2.client.secret.formula1projectdb.dfs.core.windows.net", secretid)
spark.conf.set(f"fs.azure.account.oauth2.client.endpoint.formula1projectdb.dfs.core.windows.net", 
               f"https://login.microsoftonline.com/{tenantid}/oauth2/token")
```
